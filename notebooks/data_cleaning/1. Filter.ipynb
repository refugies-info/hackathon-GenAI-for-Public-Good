```python
#@title Load CSV
import pandas as pd

df = pd.read_csv(
        "../../data/ri_annotated_texts_raw.csv",
        dtype=str,
        keep_default_na=False,
        na_values=[])
df
```

```python
#@title Remove rows where Stratégie is not N/A
df = df[df["Stratégie"] != "N/A"]

# Reindex dataframe
df = df.reset_index(drop=True)
df
```

```python
#@title Remove rows marked NONE
df = df[df["Version initiale"] != "NONE"]
df = df[df["Version retraitée"] != "NONE"]

# Reindex dataframe
df = df.reset_index(drop=True)
df
```

```python
#@title Export CSV
df.to_csv("../../data/ri_annotated_texts_filtered.csv", index=False)
```
