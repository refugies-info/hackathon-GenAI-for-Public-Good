```python
#@title Load CSV
import pandas as pd

df = pd.read_csv("../../data/ri_annotated_texts_filtered_anonymized.csv")
df
```

```python
#@title Display unique strategies

unique_strategies = df['Stratégie'].unique()
unique_strategies = sorted(unique_strategies)
unique_strategies
```

```python
#@title Map Strategies to Categories

strategies_categories = {
    'Complement': 'Complement',
    'Compression': 'Compression',
    'Examples': 'Explanation',
    'Explanation': 'Explanation',
    'Modulation': 'Modulation',
    'Omission': 'Omission',
    'SynAcronyme': 'Synonymy',
    'SynSemantic': 'Synonymy',
    'Syntactic': 'Syntactic',
    'Transcription': 'Transcription',
    'Transposition': 'Transposition'
}

df['Catégorie'] = df['Stratégie'].map(strategies_categories)
df
```

```python
#@title Cleanup Dataframe

# Remove Stratégies column
df = df[['Version initiale', 'Version retraitée', 'Catégorie']]

# Remove duplicates
df = df.drop_duplicates()
df
```

```python
#@title Save CSV

df.to_csv("../../data/ri_annotated_texts_final.csv", index=False)
```
